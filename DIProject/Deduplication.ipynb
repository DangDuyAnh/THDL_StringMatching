```python
compare_cl = recordlinkage.Compare()
# compare_cl.string("Ti√™u ƒë·ªÅ", "Ti√™u ƒë·ªÅ", label="Ti√™u ƒë·ªÅ")
compare_cl.string("Ti√™u ƒë·ªÅ", "Ti√™u ƒë·ªÅ", method="levenshtein", threshold=0.9, label="Ti√™u ƒë·ªÅ")
compare_cl.string("ƒê·ªãa ch·ªâ", "ƒê·ªãa ch·ªâ", method="levenshtein", threshold=0.9, label="ƒê·ªãa ch·ªâ")
compare_cl.string("M·ª©c gi√°", "M·ª©c gi√°", method="levenshtein", threshold=0.9, label="M·ª©c gi√°")
compare_cl.string("Di·ªán t√≠ch", "Di·ªán t√≠ch", method="levenshtein", threshold=0.9, label="Di·ªán t√≠ch")

features = compare_cl.compute(candidate_links, dataset)
```

```python
matches = features[features.sum(axis=1) > 3.6]
delete_data = dataset.iloc[matches.index.unique(1)]['Ti√™u ƒë·ªÅ']
delete_data = delete_data.reset_index() 
```
