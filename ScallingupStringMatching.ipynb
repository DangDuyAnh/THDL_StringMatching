```python
import py_stringsimjoin as ssj
import py_stringmatching as sm
import pandas as pd
```

# 1. Đọc dữ liệu

```python
hospital_accounts = pd.read_csv('data/hospital_account_info.csv')
hospital_reimbursement = pd.read_csv('data/hospital_reimbursement.csv')
print('Số bản ghi trong bảng hospital_accounts: ' + str(len(hospital_accounts)))
print('Số bản ghi trong bảng hospital_reimbursement: ' + str(len(hospital_reimbursement)))
```

# 3. Tìm các ứng cử viên

3.1 Overlap Filter

```python
OverlapFilter = ssj.OverlapFilter(sm.WhitespaceTokenizer(), overlap_size=3)
candidates = OverlapFilter.filter_tables(hospital_accounts, hospital_reimbursement, 'Account_Num', 'Provider_Num', 'Facility Name', 'Provider Name', 
['Facility Name', 'Address'], ['Provider Name', 'Provider Street Address'], out_sim_score=True)
candidates.head(10)
```

```python
print('Số lượng các cặp matching string: ' + str(len(candidates)))
```

3.2 Size Filter

```python
SizeFilter = ssj.SizeFilter(sm.WhitespaceTokenizer(), 'JACCARD', 0.9)
candidates = SizeFilter.filter_tables(hospital_accounts, hospital_reimbursement, 'Account_Num', 'Provider_Num', 'Facility Name', 'Provider Name', 
['Facility Name', 'Address'], ['Provider Name', 'Provider Street Address'])
candidates.head(10)
```

```python
print('Số lượng các cặp matching string: ' + str(len(candidates)))
```

3.3 Prefix Filter

```python
PrefixFilter = ssj.PrefixFilter(sm.WhitespaceTokenizer(),'JACCARD', 0.9)
candidates = OverlapFilter.filter_tables(hospital_accounts, hospital_reimbursement, 'Account_Num', 'Provider_Num', 'Facility Name', 'Provider Name', 
['Facility Name', 'Address'], ['Provider Name', 'Provider Street Address'])
candidates.head(10)
```

```python
print('Số lượng các cặp matching string: ' + str(len(candidates)))
```

```python
PositionFilter = ssj.PositionFilter(sm.WhitespaceTokenizer(),'JACCARD', 0.9)
candidates = PositionFilter.filter_tables(hospital_accounts, hospital_reimbursement, 'Account_Num', 'Provider_Num', 'Facility Name', 'Provider Name', 
['Facility Name', 'Address'], ['Provider Name', 'Provider Street Address'])
candidates.head(10)
```

```python
print('Số lượng các cặp matching string: ' + str(len(candidates)))
```
