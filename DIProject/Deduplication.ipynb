```python
def normalize_room(x):  
  if str(x) != 'Ch∆∞a c√≥ th√¥ng tin': 
    x = x.split(' ')
    if len(x) > 1:
      if x[0] != 'nan':
        return str(int(x[-2]))
    else:
      try:
        x = int(float(x[0]))
        return str(x)
      except:
        return str('Ch∆∞a c√≥ th√¥ng tin')
  return str('Ch∆∞a c√≥ th√¥ng tin')
```

```python
dataset['S·ªë ph√≤ng ng·ªß'] = dataset['S·ªë ph√≤ng ng·ªß'].apply(lambda x: normalize_room(str(x)))      
dataset['S·ªë toilet'] = dataset['S·ªë toilet'].apply(lambda x: normalize_room(str(x)))      
```
